```python
import PyPDF2
import pyttsx3

engine = pyttsx3.init()
voices = engine.getProperty("voices")
for i in range(len(voices)):
    print(i, voices[i])
    
engine.setProperty("voices", voices[26].id)
PDF_Reader = PyPDF2.PdfReader(open("Expos√© groupe 7 sur Chatgpt et l'API de GPT.pdf", "rb"))

for Page_num in range(len(PDF_Reader.pages)):
    Text = PDF_Reader.pages[Page_num].extract_text()
    #engine.say(Text)
    engine.save_to_file(Text, "audio.mp3")
    engine.runAndWait()
    print("end")
```
